```python
import gradio as gr
```

```python
def correct(sentence):
  crr_sntnc = gf.correct(sentence) # gramformer correct
  return list(crr_sntnc)[0]

```

```python
correct("I like for walks")
```

```python
interface = gr.Interface(fn = correct, inputs = 'text', outputs='text')
```

## the interface class requires three arguments
* fn: an ML function to run
* inputs: the input format
* outputs: the output format

```python
interface.launch(debug = True)
```
